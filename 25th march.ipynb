```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the flight price dataset
flight_data = pd.read_csv('Data_Train1.csv')

# Examine dimensions of the dataset
rows, columns = flight_data.shape
print(f"The dataset has {rows} rows and {columns} columns.")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(flight_data['Price'], bins=20, kde=True)
plt.title('Distribution of Flight Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
min_price = flight_data['Price'].min()
max_price = flight_data['Price'].max()
print(f"The range of prices in the dataset is from {min_price} to {max_price}.")
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=flight_data, x='Airline', y='Price')
plt.title('Price Variation by Airline')
plt.xlabel('Airline')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Q9. Load the Google Playstore dataset and examine its dimensions
playstore_data = pd.read_csv('googleplaystore.csv')
rows, columns = playstore_data.shape
```

```python
playstore_data.head()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Rating', data=playstore_data)
plt.xticks(rotation=90)
plt.title('Rating of Apps by Category')
plt.xlabel('Category')
plt.ylabel('Rating')
plt.show()
```

```python
missing_values = playstore_data.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(playstore_data['Size'], playstore_data['Rating'])
plt.title('Relationship between App Size and Rating')
plt.xlabel('Size')
plt.ylabel('Rating')
plt.show()
```

```python
average_prices = playstore_data.groupby('Type')['Price'].mean()
average_prices.plot(kind='bar', figsize=(8, 6))
plt.title('Average Price by App Type')
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.xticks(rotation=0)
plt.show()
```

```python
# Q14. Top 10 most popular apps - Create a frequency table
top_apps = playstore_data.groupby('App')['Installs'].sum().sort_values(ascending=False).head(10)
print("Top 10 Most Popular Apps:\n", top_apps)

# Q15. Analyzing most popular app categories
popular_categories = playstore_data.groupby('Category')['Installs'].sum().sort_values(ascending=False)
print("Most Popular App Categories:\n", popular_categories.head())
```
